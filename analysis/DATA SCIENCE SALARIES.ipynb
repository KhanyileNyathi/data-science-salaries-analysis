```python
plt.bar(df.work_year, df.salary_in_usd)
plt.title('Total salaries paid per year')
plt.grid(color='gray', linestyle='dashed')
plt.ylabel('salaries')
plt.xlabel('year')
plt.show()
```

```python
#plot the country that paid the most 
plt.figure(figsize=(25,10))
plt.bar(df['employee_residence'],df['salary_in_usd'])
plt.grid(color='gray', linestyle='dashed')
plt.title('Total salaries paid per Country')
plt.ylabel('salaries')
plt.xlabel('Country')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.bar(df['job_title'],df['salary_in_usd'])
plt.title('Total salaries paid per job')
plt.ylabel('salaries')
plt.xlabel('Job Title')
plt.grid(color='gray', linestyle='dashed')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df['experience_level'],df['salary_in_usd'])
plt.title('Total salaries paid by level of xp')
plt.ylabel('salaries')
plt.xlabel('experience level')
plt.grid(color='gray', linestyle='dashed')
plt.xticks(rotation='vertical')
plt.show()
```

FIND THE PERCENTAGE GROWTH OF THE SALARIES FROM 2020-2024


```python
#percentage growth of salaries from the year 2020-2024
pct_growth=(df.at[2024, 'salary_in_usd']-df.at[2020, 'salary_in_usd'])/df.at[2020, 'salary_in_usd']*100
pct_growth.round(0)
```
